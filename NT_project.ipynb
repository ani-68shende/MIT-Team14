```python
file = open("/Users/ani/Downloads/Stock List.csv")
file1 = pd.read_csv(file)
date = file1["symbol"]
freq = date.value_counts(ascending = True)
freq.sort_index(inplace = True)
x = freq.index
y = freq
plt.plot(x,y,marker = "o")
plt.xlabel("Months")
plt.show()
for x1 in range(8):
    print(x[x1], y[x1])
```
